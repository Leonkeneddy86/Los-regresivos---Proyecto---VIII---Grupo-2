```python
import pandas as pd

df = pd.read_csv("data/clean_data.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
# Aplicar Label Encoding a gender
label_encoder = LabelEncoder()
df['Gender_label'] = label_encoder.fit_transform(df['Gender'])
# Mostrar cómo se asignaron los valores
print("\nAsignación de valores en Label Encoding:")
for clase, valor in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)):
    print(f"{clase}: {valor}")
```

```python
# Aplicar Label Encoding a Customer Type
label_encoder = LabelEncoder()
df['Customer_Type_label'] = label_encoder.fit_transform(df['Customer Type'])
# Mostrar cómo se asignaron los valores
print("\nAsignación de valores en Label Encoding:")
for clase, valor in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)):
    print(f"{clase}: {valor}")
```

```python
# Aplicar Label Encoding a Type of Travel
label_encoder = LabelEncoder()
df['Type_of_Travel_label'] = label_encoder.fit_transform(df['Type of Travel'])
# Mostrar cómo se asignaron los valores
print("\nAsignación de valores en Label Encoding:")
for clase, valor in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)):
    print(f"{clase}: {valor}")
```

```python
# Aplicar Label Encoding a Class
label_encoder = LabelEncoder()
df['Class_label'] = label_encoder.fit_transform(df['Class'])
# Mostrar cómo se asignaron los valores
print("\nAsignación de valores en Label Encoding:")
for clase, valor in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)):
    print(f"{clase}: {valor}")
```

```python
# Aplicar Label Encoding a satisfaction
label_encoder = LabelEncoder()
df['satisfaction_label'] = label_encoder.fit_transform(df['satisfaction'])
# Mostrar cómo se asignaron los valores
print("\nAsignación de valores en Label Encoding:")
for clase, valor in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)):
    print(f"{clase}: {valor}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
import joblib

#definimos las caracteristicas y la variable objetivo
x = df[['Gender_label','Customer_Type_label','Type_of_Travel_label','Class_label','Age','Flight Distance','Inflight wifi service',
        'Departure/Arrival time convenient','Ease of Online booking','Gate location','Food and drink','Online boarding','Seat comfort'
        ,'Inflight entertainment','On-board service','Leg room service','Baggage handling','Checkin service','Inflight service',
        'Cleanliness','Departure Delay in Minutes','Arrival Delay in Minutes']]
y = df[['satisfaction_label']]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train,)

y_pred_test = model.predict(x_test)
y_pred_train = model.predict(x_train)

# Calcular Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred_test)
print(f"Mean Squared Error (MSE): {mse}")

# También puedes calcular RMSE (Root Mean Squared Error)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Calcular R² (esto es lo mismo que rf_model.score())
r2 = r2_score(y_test, y_pred_test)
print(f"R² Score: {r2}")
```

```python
from sklearn.linear_model import Ridge, Lasso

# Usar Ridge Regression para comparar con Linear Regression
model = Ridge(alpha=1.0)  
model.fit(x_train, y_train)

y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

# Calcular overfitting LinearRegression
r2_train_lr = r2_score(y_train, y_pred_train)
r2_test_lr = r2_score(y_test, y_pred_test)
overfitting_lr = (r2_train_lr - r2_test_lr) * 100

print(f"\n=== RIDGE REGRESSION ===")
print(f"R² Train: {r2_train_lr:.4f}")
print(f"R² Test: {r2_test_lr:.4f}")
print(f"Overfitting: {overfitting_lr:.2f}%")
```

```python
# Configurar RandomForest para reducir overfitting a ~5%
rf_model = RandomForestRegressor(
    n_estimators=50,           # Reducir árboles
    max_depth=8,               # Limitar profundidad
    min_samples_split=20,       # Más muestras por split
    min_samples_leaf=10,         # Más muestras por hoja
    max_features='sqrt',        # Limitar features por árbol
    random_state=42,
    n_jobs=-1
)

rf_model.fit(x_train, y_train)
y_pred = rf_model.predict(x_test)

# Calcular Mean Squared Error (MSE)
rf_mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {rf_mse}")

# También puedes calcular RMSE (Root Mean Squared Error)
rf_rmse = np.sqrt(rf_mse)
print(f"Root Mean Squared Error (RMSE): {rf_rmse}")

# Calcular R² (esto es lo mismo que rf_model.score())
rf_r2 = r2_score(y_test, y_pred)
print(f"R² Score: {rf_r2}")
```

```python
pred_df = pd.DataFrame({
    'Actual Value': y_test.values.ravel(),
    'Predicted Value': y_pred.ravel(),
    'Difference': y_test.values.ravel() - y_pred.ravel()
})
pred_df
```

```python
# 1. PRIMERO: Obtén predicciones en TRAIN (esto es lo que te falta)
y_train_pred = rf_model.predict(x_train)

# 2. Calcula R² para ambos conjuntos
from sklearn.metrics import r2_score, mean_squared_error

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_pred)

# 3. FÓRMULA PRINCIPAL para detectar overfitting:
diferencia_r2 = r2_train - r2_test
tiene_overfitting = diferencia_r2 > 0.1  # Umbral común

print(f"R² Train: {r2_train:.4f}")
print(f"R² Test: {r2_test:.4f}")
print(f"Overfitting (diferencia R²): {diferencia_r2:.4f} ({diferencia_r2*100:.2f}%)")
```

```python
# Si error en test es >50% mayor que en train → Overfitting
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred))
overfitting_rmse = ((rmse_test - rmse_train) / rmse_train) * 100

print(f"Overfitting RMSE: {overfitting_rmse:.2f}%")
```
