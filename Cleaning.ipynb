```python
import pandas as pd
```

```python
df = pd.read_csv('airline_passenger_satisfaction.csv')
```

```python
df.to_csv('airline_passenger_satisfaction.csv')
```

```python
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
df['id'] = df['id'].astype('string')
df['Gender'] = df['Gender'].astype('string')
df['Customer Type'] = df['Customer Type'].astype('string')
df['Type of Travel'] = df['Type of Travel'].astype('string')
df['Class'] = df['Class'].astype('string')
df['satisfaction'] = df['satisfaction'].astype('string')
```
